```python
import matplotlib
import math
import numpy as np
from matplotlib import pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

```python
plate_num = 'Plate00268'
line_file = f'{plate_num}/line12.5.dat'
#line_file = f'{plate_num}/line20.0.dat'

path_to_file = f'/content/drive/My Drive/TungstenQC/ThicknessData/{line_file}'
data = np.loadtxt(f'{path_to_file}')
```

```python
x = np.loadtxt(f'{path_to_file}', usecols = 1)
meas_top = np.loadtxt(f'{path_to_file}', usecols=2)
meas_btm = np.loadtxt(f'{path_to_file}', usecols=3)
th_manual = np.loadtxt(f'{path_to_file}', usecols=5)
th_calib = np.loadtxt(f'{path_to_file}', usecols=6)
```

```python
pos_top = meas_top
pos_btm = - meas_btm + th_calib - th_manual
thickness = pos_top - pos_btm
```

```python
plt.xlabel('Stage Position (cm)')  # x軸ラベル
plt.ylabel('Position (mm) ')  # y軸ラベル
plt.title(f'{plate_num}') # グラフタイトル
# ax.set_aspect('equal') # スケールを揃える
plt.grid()            # 罫線

c1,c2,c3,c4 = "blue","green","red","black"
plt.plot(x, pos_top, color=c4, label="top", marker="o")
plt.plot(x, pos_btm, color=c3, label="bottom", marker="*")
```

```python
plt.xlabel('Stage Position (cm)')  # x軸ラベル
plt.ylabel('Thickness (mm) ')  # y軸ラベル
plt.title(f'{plate_num}') # グラフタイトル
# ax.set_aspect('equal') # スケールを揃える
plt.grid()            # 罫線

plt.plot(x, thickness, color="black", label="top", marker="o")
```
